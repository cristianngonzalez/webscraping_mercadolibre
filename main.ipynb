```python
import sys
!{sys.executable} -m pip install -U selenium

from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium import webdriver
import unittest
import time

driver = webdriver.Chrome("chromedriver.exe")
driver.get('https://www.mercadolibre.com.ar');


dataset_marcas = [];
dataset_modelos = [];
dataset_anos = [];
dataset_versiones = [];
dataset_colores = [];
dataset_combustibles = [];
dataset_puertas = [];
dataset_transmiciones = [];
dataset_motores = [];
dataset_tipos = [];
dataset_kilometros = [];
dataset_ubicaciones = [];
dataset_whatsapp = [];
dataset_precios = [];
dataset = {
    'Marca: ': [], 
    'Modelo': [],
    'Ano': [],
    'Version': [],
    'Color': [],
    'Combustible': [],
    'Puertas': [],
    'Transmicion': [],
    'Motor': [],
    'Tipo': [],
    'Kilometros': [],
    'Ubicacion': [],
    'Whatsapp': [],
    'Precio': [],
}

#===================================================================================
#Buscar renault 12
time.sleep(2);
input_search = driver.find_element(By.XPATH, '//*[@id="cb1-edit"]');
input_search.click();
input_search.send_keys('Renault 12');
input_search.send_keys(Keys.ENTER);

#===================================================================================
#Recolectando los divs de resultados de productos
time.sleep(30);
divs_product = driver.find_elements(By.CLASS_NAME, "ui-search-layout__item")

#=======================================================================================
#Recorriendo los productos
for div_product in divs_product:
    #Obtener link
    link_product = div_product.find_elements(By.TAG_NAME , 'a')[0].get_attribute('href');
    #Crear la ventana con el producto
    driver.execute_script("window.open('');");
    driver.switch_to.window(driver.window_handles[1]),#Mover a la ventana nueva
    driver.get(link_product);
    time.sleep(10);
    
    try:
        #Titulo del producto
        title_product = driver.find_elements(By.CLASS_NAME, "ui-pdp-title")[0].text;
        #Tomamos la tabla del producto
        info_table_product = driver.find_elements(By.CLASS_NAME, "andes-table__body")[0];
        trs_info_table_product = info_table_product.find_elements(By.TAG_NAME , 'tr');
        
        #Loopeando las filas de la tabla
        for index , caract in enumerate( trs_info_table_product ):
            #Verificamos que caracteristica de la tabla estamos para ya asignar valores
            caract_title = caract.find_elements(By.TAG_NAME , "th")[0].text;
            caract_text = caract.find_elements(By.TAG_NAME, "span")[0].text;
            if caract_title == 'Marca':
                dataset_marcas.append(caract_text);
            elif caract_title == 'Modelo':
                dataset_modelos.append(caract_text);
            elif caract_title == 'Año':
                dataset_anos.append(caract_text);
            elif caract_title == 'Versión':
                dataset_versiones.append(caract_text);
            elif caract_title == 'Color':
                dataset_colores.append(caract_text);
            elif caract_title == 'Tipo de combustible':
                dataset_combustibles.append(caract_text);
            elif caract_title == 'Puertas':
                dataset_puertas.append(caract_text);
            elif caract_title == 'Transmisión':
                dataset_transmiciones.append(caract_text);
            elif caract_title == 'Motor':
                dataset_motores.append(caract_text);
            elif caract_title == 'Tipo de carrocería':
                dataset_tipos.append(caract_text);
            elif caract_title == 'Kilómetros':
                dataset_kilometros.append(caract_text);
        
        #Agregando la ubicacion, whatsapp y demas que estan fuera de la tabla
        dataset_precios.append( caract.find_element(By.XPATH , '//*[@id="price"]/div/div/span/span[3]' ).text );
        dataset_ubicaciones.append( caract.find_element(By.XPATH , '//*[@id="seller_profile"]/ul/div[2]/div/p' ).text );
        
    except:
        print('Paso algo aqui :(');
    else:
        print(title_product);
    finally:
        #Cerramos la ventana
        driver.close();
        #Movemos otra vez a la ventana inicial
        driver.switch_to.window(driver.window_handles[0]);
        time.sleep(3);


time.sleep(10);
driver.close()
```

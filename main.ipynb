```python
import sys
!{sys.executable} -m pip install -U selenium

import pandas as pd
import numpy as np

from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium import webdriver
import unittest
import time

driver = webdriver.Chrome("chromedriver.exe");
driver.get('https://www.mercadolibre.com.ar');


dataset_marcas = [];
dataset_modelos = [];
dataset_anos = [];
dataset_versiones = [];
dataset_colores = [];
dataset_combustibles = [];
dataset_puertas = [];
dataset_transmiciones = [];
dataset_motores = [];
dataset_tipos = [];
dataset_kilometros = [];
dataset_precios = [];


#===================================================================================
#Buscar renault 12
time.sleep(2);
input_search = driver.find_element(By.XPATH, '//*[@id="cb1-edit"]');
input_search.click();
input_search.send_keys('Renault 12');
input_search.send_keys(Keys.ENTER);

#===================================================================================
#Recolectando los divs de resultados de productos
time.sleep(30);
divs_product = driver.find_elements(By.CLASS_NAME, "ui-search-layout__item")

#=======================================================================================
#Recorriendo los productos
for div_product in divs_product:
    #Obtener link
    link_product = div_product.find_elements(By.TAG_NAME , 'a')[0].get_attribute('href');
    #Crear la ventana con el producto
    driver.execute_script("window.open('');");
    driver.switch_to.window(driver.window_handles[1]),#Mover a la ventana nueva
    driver.get(link_product);
    print('===========================================================================')
    print(link_product);
    time.sleep(10);
    
    try:
        #Titulo del producto
        title_product = driver.find_elements(By.CLASS_NAME, "ui-pdp-title")[0].text;
        #Tomamos la tabla del producto
        info_table_product = driver.find_elements(By.CLASS_NAME, "andes-table__body")[0];
        trs_info_table_product = info_table_product.find_elements(By.TAG_NAME , 'tr');
        
        marca_este_producto = 'No definido';
        modelo_este_producto = 'No definido';
        ano_este_producto = 'No definido';
        version_este_producto = 'No definido';
        color_este_producto = 'No definido';
        combustible_este_producto = 'No definido';
        puertas_este_producto = 'No definido';
        transmicion_este_producto = 'No definido';
        motor_este_producto = 'No definido';
        tipo_este_producto = 'No definido';
        kilometros_este_producto = 'No definido';
        
        #Loopeando las filas de la tabla
        for index , caract in enumerate( trs_info_table_product ):
            #Verificamos que caracteristica de la tabla estamos para ya asignar valores
            caract_title = caract.find_elements(By.TAG_NAME , "th")[0].text;
            caract_text = caract.find_elements(By.TAG_NAME, "span")[0].text;
            
            if caract_title == 'Marca':
                marca_este_producto = caract_text;
            elif caract_title == 'Modelo':
                modelo_este_producto = caract_text;
            elif caract_title == 'Año':
                ano_este_producto = caract_text;
            elif caract_title == 'Versión':
                version_este_producto = caract_text;
            elif caract_title == 'Color':
                color_este_producto = caract_text;
            elif caract_title == 'Tipo de combustible':
                combustible_este_producto = caract_text;
            elif caract_title == 'Puertas':
                puertas_este_producto = caract_text;
            elif caract_title == 'Transmisión':
                transmicion_este_producto = caract_text;
            elif caract_title == 'Motor':
                motor_este_producto = caract_text;
            elif caract_title == 'Tipo de carrocería':
                tipo_este_producto = caract_text;
            elif caract_title == 'Kilómetros':
                kilometros_este_producto = caract_text;
        
        #Agregando la ubicacion, whatsapp y demas que estan fuera de la tabla
        
        dataset_marcas.append(marca_este_producto);
        dataset_modelos.append(modelo_este_producto);
        dataset_anos.append(ano_este_producto);
        dataset_versiones.append(version_este_producto);
        dataset_colores.append(color_este_producto);
        dataset_combustibles.append(combustible_este_producto);
        dataset_puertas.append(puertas_este_producto);
        dataset_transmiciones.append(transmicion_este_producto);
        dataset_motores.append(motor_este_producto);
        dataset_tipos.append(tipo_este_producto);
        dataset_kilometros.append(kilometros_este_producto);
        dataset_precios.append( caract.find_element(By.XPATH , '//*[@id="price"]/div/div/span/span[3]' ).text );
        time.sleep(5);
        
    except:
        print('Paso algo aqui :(');
    else:
        print(title_product);
    finally:
        #Cerramos la ventana
        driver.close();
        #Movemos otra vez a la ventana inicial
        driver.switch_to.window(driver.window_handles[0]);
        time.sleep(3);
        
time.sleep(2);
dataset = {
    'Marca': dataset_marcas, 
    'Modelo': dataset_modelos,
    'Ano': dataset_anos,
    'Version': dataset_versiones,
    'Color': dataset_colores,
    'Combustible': dataset_combustibles,
    'Puertas': dataset_puertas,
    'Transmicion': dataset_transmiciones,
    'Motor': dataset_motores,
    'Tipo': dataset_tipos,
    'Kilometros': dataset_kilometros,
    'Precio': dataset_precios,
}



dataframe = pd.DataFrame(dataset);
dataframe.to_csv('renaults.csv')  

time.sleep(10);
driver.close()
```
